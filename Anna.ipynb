```python
from utils import hdf5_getters
import os
from pyspark.sql import SparkSession
from pyspark.sql import SQLContext
from pyspark.sql.functions import udf
from pyspark.sql.functions import lit
from pprint import pprint
from pyspark.sql.types import *
import time

spark_session = SparkSession\
                .builder\
                .master("local") \
                .appName("first_notebook")\
                .getOrCreate()

spark_context = spark_session.sparkContext
sqlContext = SQLContext(spark_session)
```

```python
df = spark_session.read \
    .option("header", "true")\
    .csv("songs.csv")\
    .cache()
```

```python
df.take(1)
```

```python
df = df.withColumn("loudness", df["loudness"].cast(DoubleType()))
df = df.withColumn("tempo", df["tempo"].cast(DoubleType()))
df = df.withColumn("release_year", df["release year"].cast(IntegerType()))
#df = df.withColumn("danceability", df["danceability"].cast(IntegerType()))
```

```python
df.registerTempTable("songs")

start_time_query = time.clock()
spark_session.sql("select avg(loudness), release_year from songs group by release_year order by avg(loudness) desc").show(500)
end_time_query = time.clock()
run_time_query = end_time_query - start_time_query
```

```python
df1 = df.filter(df.release_year>0)
```

```python
df_decade = df1.withColumn('decade',lit(0))
```

```python
#df_decade.show(20)
```

```python
def year_to_decade(x):
    try:
        if not x.split(',')[0] == '0':
            return d.value[int(x.split(',')[0][0:2])]
        else:
            return
    except:
        return
```

```python
udf_year_to_decade = udf(year_to_decade, StringType())
df_decade.withColumn("decade",udf_year_to_decade("decade"))
```

```python
df_decade.filter(df_decade.release_year>0).select('release_year', 'Decade').show(500)
```
